1 - Linha: Crie a classe Linha que tem dois atributos, coordenada1 e coordenada2. Cada coordenada é uma tupla que carrega duas coordenadas cartesianas (x,y) que denotam pontos do segmento de reta. Faça métodos que calculem o comprimento do segmento de reta e sua inclinação.

```python
class Linha:
    def __init__(self, coor1=(0,0), coor2=(0,0)):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def setCoor1(self,x1,y1):
        self.coor1 = (x1,y1)
        
    def setCoor2(self,x2,y2):
        self.coor2 = (x2,y2)
    
    #Calcula comprimento do segmento de reta
    def compReta(self):
        return ((self.coor2[0]-self.coor1[0])**2 + (self.coor2[1]-self.coor1[1])**2)**0.5
    
    #Calcula a inclinação
    def getInclina(self):
        return (self.coor1[1] - self.coor2[1]) / (self.coor1[0] - self.coor2[0])       

l = Linha()
print("Digite os valores da primeira coordenada")
l.setCoor1(int(input("Digite o valor de X ")), int(input("Digite o valor de Y ")))
print("Digite os valores da segunda coordenada")
l.setCoor2(int(input("Digite o valor de X ")), int(input("Digite o valor de Y ")))
print(f"Coordenada 1 = {l.coor1}\nCoordenada 2 = {l.coor2}")
print(f"O comprimento do segmento de reta é: {l.compReta():.2f}")
print(f"A inclinação é = {l.getInclina():.2f}")
```

```python
import abc
import math

class Figura:
    def __init__(self):
        print("Figura criada")
    
    def getArea(self):
        pass
    
    def getPerimetro(self):
        pass

class Poligono(Figura):
    def __init__(self, base=1, altura=1):
        Figura.__init__(self)
        self.base = base
        self.altura = altura
        
    def getArea(self):
        return self.base * self.altura
    
class Diagonal(abc.ABC):
    @abc.abstractmethod
    def getDiagonal(self, base, altura):
        return math.sqrt((self.base**2) + (self.altura**2))
    

class Triangulo(Poligono):
    def __init__(self, lado1=1, lado2=1):
        Poligono.__init__(self, base=1, altura=1)
        self.base = float(input("Digite a base do triângulo: "))
        self.altura = float(input("Digite a altura do triângulo: "))
        self.lado1 = float(input("Digite um dos lados do triângulo: "))
        self.lado2 = float(input("Digite o outro lado do triângulo: "))
        print("Triângulo criado")
    
    def getArea(self):
        return Poligono.getArea(self)/2
    
    def getPerimetro(self):
        return self.lado1 + self.lado2 + self.base
        
class Retangulo(Poligono, Diagonal):
    def __init__(self):
        Poligono.__init__(self, base=1, altura=1)
        self.base = float(input("Digite a base do Retângulo: "))
        self.altura = float(input("Digite a altura do Retângulo: "))
        print("Retângulo criado")
        
    def getDiagonal(self, base, altura):
        base = self.base
        altura = self.altura
        return Diagonal.getDiagonal(self, base, altura)
    
    def getArea(self):
        return Poligono.getArea(self)
    
    def getPerimetro(self):
        return 2 * (self.base + self.altura)
        
class Quadrado(Poligono, Diagonal):
    def __init__(self):
        Poligono.__init__(self, base=1, altura=1)
        self.base = float(input("Digite o valor dos lados do Quadrado: "))
        self.altura = self.base
        print("Quadrado criado")
        
    def getDiagonal(self, base, altura):
        base = self.base
        altura = self.altura
        return Diagonal.getDiagonal(self, base, altura)
    
    def getArea(self):
        return Poligono.getArea(self)
    
    def getPerimetro(self):
        return 2 * (self.base + self.altura)
    
class Losango(Poligono):
    def __init__(self, angM=1, angm=1):
        Poligono.__init__(self, base=1, altura=1)
        self.base = float(input("Digite o valor dos lados do Losângo: "))
        self.altura = self.base
        self.angM = float(input("Digite o ângulo maior: "))
        self.angm = (360-(2*self.angM)) / 2
        print("Losango criado")
        
    def getDiagonalMaior(self):
        return math.sqrt((self.base**2) + (self.altura**2) - (2 * self.base * self.altura * math.cos(converterGrausParaRad(self.angM))))
    
    def getDiagonalMenor(self):
        return math.sqrt(self.base**2 + self.altura**2 - 2 * self.base * self.altura * math.cos(converterGrausParaRad(self.angm)))
    
    def getArea(self):
        return (self.getDiagonalMaior() * self.getDiagonalMenor())/2
    
    def getPerimetro(self):
        return 2 * (self.base + self.altura)
    
    
def converterGrausParaRad(numero):
        rad = (numero/180)*math.pi
        return rad
    
class Circulo(Figura):
    def __init__(self, radius=0):
        Figura.__init__(self)
        self.radius = float(input("Digite um raio para o círculo: "))
        print("Círculo criado")
        
    def getArea(self):
        return math.pi*(self.radius**2)
    
    def getPerimetro(self):
        return 2*math.pi*self.radius
        
        
class Geometria():
    c = Circulo()
    t = Triangulo()
    q = Quadrado()
    r = Retangulo()
    los = Losango()
    lista = [c,t,q,r,l]
    print(
        f"--Círculo--\n  Raio: {c.radius:.2f}\n  Área: {c.getArea():.2f}\n  Perímetro: {c.getPerimetro():.2f}"
    )
    print(
        f"--Triângulo--\n  Lado 1: {t.lado1} / Lado 2 : {t.lado2} / Base: {t.base} / Altura: {t.altura}\n  Área: {t.getArea():.2f}\n  Perímetro: {t.getPerimetro():.2f}"
    )
    print(
        f"--Retângulo--\n  Base: {r.base} / Altura: {r.altura}\n  Área = {r.getArea():.2f}\n  Perímetro = {r.getPerimetro():.2f}\n  Diagonal = {r.getDiagonal(r.base,r.altura):.2f}"
    )
    print(
        f"--Quadrado--\n  Base: {q.base} / Altura: {q.altura}\n  Área = {q.getArea():.2f}\n  Perímetro = {q.getPerimetro():.2f}\n  Diagonal = {q.getDiagonal(q.base,q.altura):.2f}"
    )
    print(
            f"--Losango--\n  Base: {los.base} / Altura: {los.altura} / Ângulo Maior: {los.angM} / Ângulo Menor: {los.angm}\n  Área = {los.getArea():.2f}\n  Perímetro = {los.getPerimetro():.2f}\n  Diagonal Maior = {los.getDiagonalMaior():.2f}\n  Diagonal Menor = {los.getDiagonalMenor():.2f}"
    )
```
